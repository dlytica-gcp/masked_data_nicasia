```python
df_300 = df.sample(n=300,random_state = 42).reset_index(drop=True)
df_300.head()
# df_300.shape
```

```python
df_300 = df_300.drop(['_airbyte_ab_id', '_airbyte_emitted_at',"source_file_path","_airbyte_additional_properties"], axis=1)
df_300.head()
```

```python
df["account_owner_cbs_id"].value_counts()
```

```python
cif_masked = pd.read_csv("../../crmuser/account/cif_masked.csv")
cif_masked.head() 
```

```python
# acid_foracid = pd.read_csv("../../tbaadm/general_acct_mast_table/acid_foracid_masked.csv")
# acid_foracid.head()
```

```python
customer_data_dln = cif_masked[cif_masked["cif_id"].str.startswith("DLNIND")]
customer_data_dln.head(10)
```

```python
df_300["account_owner_cbs_id"] = customer_data_dln["cif_id"].sample(n=len(df_300), replace=True).values
df_300.head()
```

```python
df_grouped = df_300.groupby("account_owner_cbs_id").size().reset_index(name='count').sort_values('count', ascending=False)
df_grouped.head(20)
```

```python
acid_foracid = pd.read_csv("../../tbaadm/general_acct_mast_table/acid_foracid_masked.csv")
acid_foracid.head()
```

```python
acid_foracid = acid_foracid.drop_duplicates(subset=["cif_id"])
acid_foracid.shape
```

```python
result_df = pd.merge(df_300, acid_foracid, left_on="account_owner_cbs_id", right_on="cif_id", how="left",suffixes=('_l','_r'))
print(len(result_df))
```

```python
cif_masked = pd.read_csv("../../crmuser/account/cif_masked.csv")
cif_masked.head()
```

```python
df_300["account_number"] = result_df["foracid_t"].values
```

```python
result_df1 = pd.merge(df_300, cif_masked, left_on="account_owner_cbs_id", right_on="cif_id", how="left",suffixes=('_l','_r'))
print(len(result_df1))
```

```python
df_300["account_holder_name"] = result_df1["preferredname"].values
df_300[["account_holder_name","account_number","account_owner_cbs_id"]].head()
```
