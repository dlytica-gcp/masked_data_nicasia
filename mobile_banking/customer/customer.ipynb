```python
df_114 = df[:113]
df_114.head()
```

```python
cif_masked = pd.read_csv("../../crmuser/account/cif_masked.csv")
cif_masked.head()
```

```python
customer_data_dln = cif_masked[cif_masked["cif_id"].str.startswith("DLNIND")]
customer_data_dln.head(10)
```
