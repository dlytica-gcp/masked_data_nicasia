```python
df = df.drop(['_airbyte_ab_id', '_airbyte_emitted_at',"source_file_path","_airbyte_additional_properties"], axis=1)
df.head()
```

```python
import hashlib
import base64

def custom_hash_with_prefix(password):
    # Ensure password is a string
    if not isinstance(password, str):
        password = str(password)
    # Extract first 4 chars (or fewer if password shorter)
    prefix = password[:4]
    
    # Generate SHA1 digest
    sha1_digest = hashlib.sha1(password.encode('utf-8')).digest()
    
    # Encode digest to Base64 string
    hash_b64 = base64.b64encode(sha1_digest).decode('utf-8')
    
    # Join prefix + hash
    return prefix + hash_b64

df["password"] = df["password"].apply(custom_hash_with_prefix)
df["txn_password"] = df["txn_password"].apply(custom_hash_with_prefix)
```
