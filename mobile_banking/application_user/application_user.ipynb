```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("application_user.xlsx")
df.head() 
```

```python
df = df.drop(['_airbyte_ab_id', '_airbyte_emitted_at',"source_file_path","_airbyte_additional_properties"], axis=1)
print(df.shape)
df.head()
```

```python
import hashlib

def generate_hash(data):
    if not isinstance(data, str):
        data = str(data)
    return data[:8] +hashlib.sha1(data.encode('utf-8')).hexdigest().upper()

df["password"] = df["password"].apply(generate_hash)
```

```python
df.to_csv("application_user_masked.csv", index=False)
```
