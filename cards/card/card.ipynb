```python
df_acc= pd.read_csv("../account/account_masked.csv")
df_acc.head()
df_acc = df_acc.drop_duplicates(subset=['acc_cus_code'])
print(len(df_acc))
```

```python
result_df = pd.merge(df, df_acc, how='left', left_on="car_cus_code", right_on="acc_cus_code", suffixes=('_p', '_a'))
result_df.head()
```

```python
df_acid = pd.read_csv("../../tbaadm/general_acct_mast_table/acid_foracid_masked.csv")
df_acid.head()
```

```python
result_df = result_df.merge(df_acid, how='left', left_on="acc_numb", right_on="foracid_t", suffixes=('_p', '_a'))
result_df.head()
```

```python
df_crmuser = pd.read_csv("../../crmuser/account/cif_masked.csv")
df_crmuser.head()
```

```python
result_df = result_df.merge(df_crmuser, how='left', on="cif_id", suffixes=('_p', '_a'))
result_df.head()
```
