```python
df_atm_num_masked = pd.read_csv("atm_num_masked.csv")
res = df.merge(df_atm_num_masked, on="ate_num", how="left",suffixes=("_p", "_x"))
df['ate_num'] = res['ate_num_t'].values
# df = df.drop(['ate_num_t'], axis=1)
df.head()
```

```python
df['ate_adr1']=df['ate_adr1'].astype(str).apply(lambda x: 'Dlytica' +x[4:])
df['ate_labe']=df['ate_adr1']

df[['ate_adr1','ate_labe']].head().reset_index()
```

```python
def update_ip(ip):
    if pd.notna(ip):
        parts = ip.split('.')
        if len(parts) == 4:
            if int(parts[1]) > 4 :
                parts[1] = str(int(parts[1]) - 3)
            else:
                parts[1] = str(int(parts[1]) + 3)

            parts[2] = str(int(parts[2]) + 5)
            parts[-1] = str(int(parts[-1]) + 1)
            return '.'.join(parts)
    return ip

# Apply the function to the 'ate_ip_addr' column
df['ate_ip_addr'] = df['ate_ip_addr'].apply(update_ip)
df.head()
```

```python
df.to_csv('atm_terminal_masked.csv', index=False)
```
