```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_excel("customer_cardholder.xlsx")
df.head()
# print(len(df))
```

```python
df = df.drop(['_airbyte_ab_id', '_airbyte_emitted_at',"source_file_path","_airbyte_additional_properties"], axis=1)
```

```python
df_acc= pd.read_csv("../account/account_masked.csv")
df_acc.head()
print(len(df_acc))
df_acc = df_acc.drop_duplicates(subset=['acc_cus_code'])
print(len(df_acc))
```

```python
result_df = pd.merge(df, df_acc, how='left', left_on="cus_code", right_on="acc_cus_code", suffixes=('_p', '_a'))
result_df["acc_numb"].head()
print(len(result_df))
```

```python
df_acid = pd.read_csv("../../tbaadm/general_acct_mast_table/acid_foracid_masked.csv")
df_acid.head()
```

```python
result_df = result_df.merge(df_acid, how='left', left_on="acc_numb", right_on="foracid_t", suffixes=('_p', '_a'))
result_df.head()
```

```python
df_crmuser = pd.read_csv("../../crmuser/account/cif_masked.csv")
df_crmuser.head()
```

```python
result_df = result_df.merge(df_crmuser, how='left', on="cif_id", suffixes=('_p', '_a'))
result_df.head()
```

```python
df["cus_firs_nam1"] = result_df["cust_first_name"].values
df["cus_midd_nam1"] = result_df["cust_middle_name"].values
df["cus_last_nam1"] = result_df["cust_last_name"].values  
df["cus_email_adr"] = result_df["email"].values
df["cus_phon_numb"] = result_df["preferredphone"].values
```
