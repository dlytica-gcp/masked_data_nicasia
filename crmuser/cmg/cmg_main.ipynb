```python
import pandas as pd
```

```python
df_cmg= pd.read_excel("cmg.xlsx")
df_cmg.head()
```

```python
df_cmg = df_cmg.drop(['_airbyte_ab_id', '_airbyte_emitted_at',"source_file_path","_airbyte_additional_properties"], axis=1)

df_cmg.head()
 
```

```python
df_cmg.rename(columns={'cif_id':'orgkey'},inplace=True)
df_cmg.head()
```

```python
df_cif=pd.read_csv(r"..\account\cif_masked.csv")
df_cif.head()
```

```python
df_result = pd.merge(df_cmg, df_cif, how='left', on='orgkey',suffixes=('_p', '_a'))

df_result['cif_id'].head(5)
 
```

```python
df_account=pd.read_csv("../account/account_masked.csv")
```

```python
df_result = pd.merge(df_result, df_account, how='left', on='cif_id',suffixes=('_p', '_a'))
df_result['cif_id'].head(5)
```

```python
print(df_cmg.shape)
print(df_result.shape)
```

```python
df_cmg.sample(10)
```

```python
df_cmg['orgkey']=df_result['cif_id'].values
df_cmg['email']=df_result['preferredemail_a'].values 
df_cmg['phone']=df_result['preferredphone_a'].values 
df_cmg['corp_id']=df_result['corp_id_a'].values 
df_cmg['cust_sex']=df_result['gender_a'].values 
df_cmg['cust_id']=df_result['core_cust_id'].values
df_cmg['cust_name']=df_result['name_a'].values 
df_cmg['pan_gir_num']=df_result['pan_a'].values 
df_cmg['ichg_userid']=df_result['bomodifiedby'].values
df_cmg['rcre_user_id']=df_result['bocreatedby'].values
df_cmg['date_of_birth']=df_result['cust_dob'].values
df_cmg['cust_hlth_code']=df_result['cust_health_ref_code'].values
df_cmg['cust_first_name']=df_result['cust_first_name_a'].values 
df_cmg['cust_middle_name']=df_result['cust_middle_name_a'].values 
df_cmg['cust_last_name']=df_result['cust_last_name_a'].values 
df_cmg['cust_occp_code']=df_result['occupation'].values
df_cmg['nat_id_card_num']=df_result['nat_id_card_num_a'].values 
df_cmg['cust_introd_name']=df_result['introducername'].values
df_cmg['cust_introd_cust_id']=df_result['introducerid'].values
df_cmg['bank_id']=df_result['bank_id_a'].values 
```

```python
df_cmg.rename(columns={'orgkey':'cif_id'},inplace=True)
```
