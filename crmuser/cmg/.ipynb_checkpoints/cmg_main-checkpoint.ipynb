```python
import pandas as pd
```

```python
df_cmg= pd.read_excel("cmg.xlsx")
df_cmg.head()
```

```python
df_cmg = df_cmg.drop(['_airbyte_ab_id', '_airbyte_emitted_at',"source_file_path","_airbyte_additional_properties"], axis=1)

df_cmg.head()
 
```

```python
df_cmg.rename(columns={'cif_id':'orgkey'},inplace=True)
df_cmg.head()
```

```python
df_account=pd.read_csv(r"..\account\account_masked.csv")
df_account.head()
```

```python
df_result = pd.merge(df_cmg, df_account, how='left', on='orgkey',suffixes=('_p', '_a'))

df_result['orgkey'].head(5)
 
```
