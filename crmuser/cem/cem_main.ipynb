```python
import pandas as pd
df_cem= pd.read_excel("cem.xlsx").head()

```

```python
df_cem = df_cem.drop(['_airbyte_ab_id', '_airbyte_emitted_at',"source_file_path","_airbyte_additional_properties"], axis=1)
df_cem.head()
```

```python
df_cif = pd.read_csv("../account/cif_masked.csv")
df_cif.head()
```

```python
result_df = df_cem.merge(df_cif, how='left', left_on='cust_id', right_on="orgkey", suffixes=('_p', '_a'))
result_df.head()
```

```python
df_cem["cust_id"] = result_df["cif_id"].values
```
