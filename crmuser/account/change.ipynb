```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("account (1).csv")
df.head()
```

```python
occupation_descriptions = [
    "ENGINEER",
    "BUSINESSMAN",
    "TEACHER",
    "HOUSEHOLD",
    "SERVICE",
    "UNEMPLOYED",
    "RETIRED",
    "BANKER",
    "DOCTOR",
    "FARMER",
    "TRADING",
    "SELFEMPLOYED"
]
```

```python
df["city"].head()
```

```python
df[["preferredphone", "preferredemail","occupation"]].head()
```

```python
df['occupation'] = np.where(
    (df['occupation'].isna()) & (df['cust_type'] == 'INDIVIDUAL'),
    np.random.choice(occupation_descriptions, size=len(df)),
    df['occupation']
)
```

```python
df.to_csv('account_masked3.csv', index=False) 
```
