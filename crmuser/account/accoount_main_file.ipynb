```python
def get_occupation(data):
    if data =="NOT A":
        return "NOT AVAILABLE"
    elif data =="ENGIN":
        return "ENGINEER"
    elif data =="BUSIN":
        return "BUSINESSMAN"
    elif data =="TEACH":
        return "TEACHER"
    elif data =="HOUSE":
        return "HOUSEHOLD"
    elif data =="SERVI":
        return "SERVICE"
    elif data =="UNEMP":
        return "UNEMPLOYED"
    elif data =="RETIR":
        return "RETIRED"
    elif data =="BANKE":
        return "BANKER"
    elif data =="SERVI":
        return "SERVICE"
    elif data =="DOCTO":
        return "DOCTOR"
    elif data =="FARME":
        return "FARMER"
    elif data =="TRADI":
        return "TRADING"
    elif data =="SELFE":
        return "SELFEMPLOYED"
    
    else:
        return "nan" 

df["occupation"] =  df["occupation"].apply(get_occupation)
```

```python
def pan(data):
    random_digits = np.random.randint(100000000, 999999999)  
    return "PAN" + str(random_digits)

df["pan"] = df["pan"].apply(pan)

```

```python
#cust_type generation
import numpy as np
def get_cust_type(df):
    df= df.copy()
    condition =  (df["constitution_code"] == "INDIJ") | (df["constitution_code"] == "INDIV")
    df['cust_type'] = np.where(condition,"INDIVIDUAL","LEGAL")
    return df


df = get_cust_type(df)
df[["cust_type","constitution_code"]].head(7)
```

```python
# CIF_ID Generation
import pandas as pd
import numpy as np

def generate_cif_id(code):
    if code == "INDIVIDUAL":
        prefix = "DLNIND"
    elif code == "LEGAL":
        prefix = "DLNLEG"
    else:
        return ""  
    
    random_digits = str(np.random.randint(1000000, 9999999))
    return prefix + random_digits


df["cif_id"] = df["cust_type"].apply(generate_cif_id)
df["core_cust_id"] = df["cif_id"]
df["group_id"] = df["cif_id"]
df["introducerid"] = ""

df.head()
```

```python
import numpy as np

def generate_email(data):
    # Extract first name (handle NaN and company names)
    first_name = str(data["cust_first_name"]) if not pd.isna(data["cust_first_name"]) else "unknown"
    
    # Remove "PVT. LTD." or "LTD." and clean the name
    name = (
        first_name.replace("PVT. LTD.", "")
                 .replace("LTD.", "")
                 .strip() 
                 .lower()  
                 .replace(" ", "")
    )
    
    num = str(np.random.randint(100, 999))
    
    if data["cust_type"] == "INDIVIDUAL":
        email = f"{name}{num}@gmail.com"
    else:
        email = f"{name}{num}@outlook.com"
    
    return email

# Apply the function to the DataFrame
shuffled["email"] = shuffled[["cust_first_name", "cust_type"]].apply(generate_email, axis=1)

# Display first 15 generated emails
shuffled["email"].head(15)
```
