```python
import pandas as pd
```

```python
df = pd.read_excel('transaction_details.xlsx')
df.head()
```

```python
df = df.drop(["source_file_path"], axis=1)
df.head()
```

```python
print(len(df))
```

```python
df_mecch = pd.read_csv('../merchant_details/merchant_details_masked.csv')
df_mecch.head()
```

```python
result =  pd.merge(df, df_mecch, left_on='merchant_id',right_on="_id" ,how='left',suffixes=('_a', '_b'))
print(len(result))
```

```python
result.head(5)
```

```python
print(len(result))
```

```python
df["merchant_name"] = result["name"].values
df["terminal_name"] = result["name"].values
df["pan_number"] = result["pan_number_a"].values
df["submerchant_details_name"] = result["name"].values
df["account_number"] = result["account_number_a"].values
df["submerchant_name"] = result["name"].values
```

```python
df.to_csv('transaction_details_masked.csv', index=False)
```
