```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('c_relparty_details.xlsx')
df.head()
```

```python
df = df.drop(['_airbyte_ab_id', '_airbyte_emitted_at',"source_file_path","_airbyte_additional_properties"], axis=1)
df["foracid"].head()
```

```python
df_gam = pd.read_csv("../../tbaadm/general_acct_mast_table/acid_foracid_masked.csv")
df_gam.head()
```

```python
result_df = df.merge(
    df_gam,
    on='foracid',
    how='left'  
)

print(f"Original df length: {len(df)}")
print(f"Merged result length: {len(result_df)}")
```

```python
df["foracid"] = result_df["foracid_t"].values
```

```python
df["foracid"].head()
```

```python
df_acc = pd.read_csv("../../crmuser/account/account_masked.csv")
# df_kyc = pd.read_csv("../../custom/kyc_details/kyc_details_masked.csv")

df_acc["occupation"].head(40)
```
