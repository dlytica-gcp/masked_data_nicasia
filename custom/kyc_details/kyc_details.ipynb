```python
df_shuffle = pd.read_csv("../../crmuser/account/shuffled_data.csv")

df_shuffle.head()
```

```python
def get_surname(row):
    """Extract surname from preferredname and combine with first part of grd_name"""
    
    # Check conditions for this single row
    is_male_no_spouse = (row["gender"] == "M") and pd.isna(row["spouse_name"])
    is_female_no_spouse = (row["gender"] == "F") and pd.isna(row["spouse_name"])
    is_male_yes_spouse = (row["gender"] == "M") and pd.notna(row["spouse_name"])
    is_female_yes_spouse = (row["gender"] == "F") and pd.notna(row["spouse_name"])
    
    # Extract name components safely
    grd_name = row['grd_name'] if pd.notna(row['grd_name']) else ""
    father_name = row['father_name'] if pd.notna(row['father_name']) else ""
    preferredname = row['preferredname'] if pd.notna(row['preferredname']) else ""
    
    # Split names safely
    son_first_part = df_shuffle[df_shuffle["gender"] == "M"].sample(n=1).iloc[0]["cust_first_name"]
    daught_first_part = df_shuffle[df_shuffle["gender"] == "F"].sample(n=1).iloc[0]["cust_first_name"]
    grd_first_part = grd_name.split()[0] if grd_name and ' ' in grd_name else grd_name
    father_first_part = father_name.split()[0] if father_name and ' ' in father_name else father_name
    preferred_surname = preferredname.split()[-1] if preferredname and ' ' in preferredname else preferredname
    
    # Update common fields
    row["father_name"] = f"{father_first_part} {preferred_surname}".strip()
    row['grd_name'] = f"{grd_first_part} {preferred_surname}".strip()
    
    # Handle cases based on marital status
    if is_male_no_spouse or is_female_no_spouse:
        # Unmarried - clear child-related fields
        row["childdate1"] = ""
        row["childdate2"] = ""
        row["childname1"] = ""
        row["childname2"] = ""
        row["daught_name"] = ""
        row["son_name"] = ""
        
    elif is_male_yes_spouse or is_female_yes_spouse:
        # Married - keep existing child data or set to empty
        row["childdate1"] = row["childdate1"] if pd.notna(row["childdate1"]) else ""
        row["childdate2"] = row["childdate2"] if pd.notna(row["childdate2"]) else ""
        row["childname1"] = row["childname1"] if pd.notna(row["childname1"]) else ""
        row["childname2"] = row["childname2"] if pd.notna(row["childname2"]) else ""
        if row["daught_name"]!= "":
            row["daught_name"] = f"{daught_first_part} {preferred_surname}".strip()
        if row["son_name"] != "":
            row["son_name"] = f"{son_first_part} {preferred_surname}".strip()
        
        # Assign spouse name from shuffled data if spouse_name is null
        if pd.isna(row["spouse_name"]) or row["spouse_name"] == "":
            if row["gender"] == "M":
                # Male gets female spouse
                data = df_shuffle[df_shuffle["gender"] == "F"].sample(n=1).iloc[0]
                row["spouse_name"] = f"{data['cust_first_name']} {data['cust_last_name']}"
            elif row["gender"] == "F":
                # Female gets male spouse
                data = df_shuffle[df_shuffle["gender"] == "M"].sample(n=1).iloc[0]
                row["spouse_name"] = f"{data['cust_first_name']} {data['cust_last_name']}"
    
    return row

print("Processing all rows...")
final_df = result_df.apply(get_surname, axis=1)
print("Processing complete!")
print(f"Final dataframe shape: {final_df.shape}")

# Show sample results
print("\nSample results:")
print(final_df[["preferredname","gender", "grd_name", "father_name", "spouse_name"]].head())
final_df.head()
```

```python
df["spouse_name"] = final_df["spouse_name"].values
df["grd_name"] = final_df["grd_name"].values
df["father_name"] = final_df["father_name"].values
df["son_name"] = final_df["son_name"].values    
df["daught_name"] = final_df["daught_name"].values
df["childdate1"] = final_df["childdate1"].values
df["childdate2"] = final_df["childdate2"].values
df["childname1"] = final_df["childname1"].values
df["childname2"] = final_df["childname2"].values
df["panno"] = result_df["pan"].values
df["cif_id"] = result_df["cif_id_a"].values

```

```python
def generate_old_citizenship_number(row=None):
    unique_num = f"{np.random.randint(10000, 99999)}" 
    unque_1 = np.random.randint(1000, 9999)  
    return f"{unique_num}/{unque_1}"

df["citizenship_number"] = final_df.apply(generate_old_citizenship_number, axis=1)
```
