```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('products.xlsx')
df.head()
```

```python
df = df.drop(["source_file_path"], axis=1)
```

```python
random_codes = np.random.permutation(np.arange(1, len(df) + 1))

df["schm_code_t"] = ["PDSC" + str(num).zfill(2) for num in random_codes]

df.head()
```

```python
df[["schm_code_t", "Schm_code","Schm_Desc (Product)"]].to_csv("schm_masked.csv", index=False)
```

```python
df["Schm_code"] = df["schm_code_t"]
df.drop(["schm_code_t"], axis=1, inplace=True)
df.head()
```
